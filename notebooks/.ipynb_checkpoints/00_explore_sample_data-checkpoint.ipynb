# Define alleleFrequency()

Codes are in the form: A:T:C:G:N:del

So, if an allele is in base `T`, with code `0:155:0:75:0:0`, then 75 known mutations have occurred to position `G`.
Allele frequency, then, is the `# mutations / total`.

```python
# Allele count format: [A : T : C : G : N : del]
base_dict = {
    'A': 0,
    'T': 1,
    'C': 2,
    'G': 3,
    'N': 4,   # unused
    'del': 5  # unsued, both defined to be explicit
}

# Mutation Frequency = mutations / total
def alleleFrequency(base, allele_count):    
    # Handle empty/null error
    # if pd.isna(row[allele_count]) or pd.isna(row['base']):
    #     return 0.0

    base_index = base_dict.get(base)
    vector = [int(num) for num in allele_count.split(':')]

    # Sum all values
    total = sum(vector)

    # Error-handle for division-by-zero
    if total == 0:
        return 0.0

    # Find difference between base & total
    base_value = vector[base_index]
    mutations = total - base_value
    freq = mutations / total
    return freq

# Examples
print("Unit test cases")

freq = alleleFrequency(base="T", allele_count="0:200:0:1:0:0")
print(f"{freq}")

freq = alleleFrequency(base="C", allele_count="0:0:100:7:0:0")
print(f"{freq}")

freq = alleleFrequency(base="A", allele_count="40:1:2:3:0:0")
print(f"{freq}")
```

# Reorganise data for ML

```python
id_cols = ['chromosome', 'position', 'base']
df_subset = df.head(10_000)
pop_cols = [c for c in df_subset.columns if c.startswith('Dsim_Fl_')]
long_df = df_subset[id_cols + pop_cols].melt(
    id_vars=id_cols,
    value_vars=pop_cols,
    var_name='population',
    value_name='allele_count'
)

# Extract scenario, F value, and replicate ID
long_df[['pop_scenario', 'pop_F', 'pop_id']] = (
    long_df['population']
    .str.extract(r'^Dsim_Fl_(Base|Hot)(?:_(F\d+))?_(\d+)$')
)

# Base populations have no F value in the name
long_df['pop_F'] = long_df['pop_F'].fillna('Base')

# Final column order
long_df = long_df[
    [
        'chromosome',
        'position',
        'base',
        'pop_id',
        'pop_F',
        'pop_scenario',
        'allele_count'
    ]
]

long_df['allele_freq'] = long_df.apply(
    lambda row: alleleFrequency(
        base=row['base'],
        allele_count=row['allele_count']
    ),
    axis=1
)

long_df.drop(inplace=True, columns=['pop_scenario'])
# Sort within each population/chromosome
long_df = long_df.sort_values(
    ['chromosome', 'pop_id', 'pop_F', 'position']
)

group_cols = ['chromosome', 'pop_id', 'pop_F']

# Previous SNP
long_df['prev_position'] = (
    long_df.groupby(group_cols)['position']
    .shift(1)
)

long_df['prev_allele_freq'] = (
    long_df.groupby(group_cols)['allele_freq']
    .shift(1)
)

long_df['prev_position_diff'] = (
    long_df['position'] - long_df['prev_position']
)

# Next SNP
long_df['next_position'] = (
    long_df.groupby(group_cols)['position']
    .shift(-1)
)

long_df['next_allele_freq'] = (
    long_df.groupby(group_cols)['allele_freq']
    .shift(-1)
)

long_df['next_position_diff'] = (
    long_df['next_position'] - long_df['position']
)

long_df
```
